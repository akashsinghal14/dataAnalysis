```python
data = pd.read_csv('/kaggle/input/sales-data-for-economic-data-analysis/salesforcourse-4fe2kehu.csv')
df = data.copy()
df.head()
```

```python
# from above we can see there are 34867 rows and 16 columns
# in last column Coloumn1 has more nulls, lets drop this column
df.drop(['Column1'],axis= 1)
df.drop(['index','Column1'],axis= 1, inplace=True)
```

```python
df
```

```python
# the last row is null, lets remove it as well
df = df.dropna()
df.tail()
```

```python
df.shape
```

```python
df['Profit'] = df['Revenue'] - df['Cost']
df['Profit Margins'] = round(df['Profit']/df['Revenue'] * 100, 2)
df['Unit Profit']=df['Unit Price']-df['Unit Cost']
df['Unit Profit Margin']= round(df['Unit Profit']/df['Unit Price'] * 100 , 2)
df
```

```python
# lets see which state has more profit across all year
df.groupby(['State'])['Profit'].sum().sort_values(ascending=False)
```

```python
df.iloc[:,1] = df.Year.astype('int')
df
```

```python
df2015 = df[df.Year== 2015]
df2015.Year.describe()
```

```python
df2016 = df[df.Year== 2016]
df2016.Year.describe()
```

```python
# lets see which state has more profit n each year
df2016.groupby(['State'])['Profit'].sum().sort_values(ascending=False)
```

```python
df2015.groupby(['State'])['Profit'].sum().sort_values(ascending=False)
```

```python
# do the same above thing in one statement of code
df[df.Year==2015].groupby(['State'])['Profit'].sum().sort_values(ascending=False)
```

```python
# to do the analysis of age, I wanted to make groups of ages lets see age describe
df['Customer Age'].describe()
# 0 to 20
# 20 to 40
# 40 to 60
# >60
```

```python
for i in df["Customer Age"]:
    if i in range(15,19):
        df.loc[df["Customer Age"]== i, "Age Category"]= "15-19"
    elif i in range(20,30):
        df.loc[df["Customer Age"]== i, "Age Category"]= "20s"
    elif i in range(30,40):
        df.loc[df["Customer Age"]== i, "Age Category"]= "30s"
    elif i in range(40,50):
        df.loc[df["Customer Age"]== i, "Age Category"]= "40s"
    elif i in range(50,60):
        df.loc[df["Customer Age"]== i, "Age Category"]= "50s"
    elif i in range(60,70):
        df.loc[df["Customer Age"]== i, "Age Category"]= "60s" 
    elif i in range(70,80):
        df.loc[df["Customer Age"]== i, "Age Category"]= "70s"
    elif i in range(80,90):
        df.loc[df["Customer Age"]== i, "Age Category"]= "80s"
```
